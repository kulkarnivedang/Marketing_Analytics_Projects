```python
# The dataset

dataset = [['milk','bread', 'biscuit'],['bread','milk','biscuit','cereal'],['bread','tea'],['jam','bread','milk']
,['tea','biscuit'],['bread','tea'],['tea','cereal'],['bread','tea','biscuit'],['jam','bread','tea'],['bread','milk']
,['coffee','orange','biscuit','cereal'],['coffee','orange','biscuit','cereal'],['coffee','sugar']
,['bread','coffee','orange'],['bread','sugar','biscuit'],['coffee','sugar','cereal'],['bread','sugar','biscuit']
,['bread','coffee','sugar'],['bread','coffee','sugar'],['tea','milk','coffee','cereal']]
```

```python
dataset
```

```python
# Importing required library for market basket analysis
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
# Fitting and tranforming the dataset
data = te.fit(dataset).transform(dataset)
```

```python
data
```

```python
# Making a dataframe for further analysis
import pandas

productdata = pandas.DataFrame(data, columns=te.columns_)
```

```python
productdata
```

```python
# Importing apriori and association rule for Market Basket Analysis

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Determine the frequent factors
frequent_factors = apriori(productdata, use_colnames=True, min_support = 0.01, max_len=3)
```

```python
# Determine the rules
rules = association_rules(frequent_factors, metric='lift', min_threshold=1)
```

```python
print(rules.head(10).sort_values(by='confidence', ascending=False))
```

```python
# Determine product with the maximum support
apriori(productdata, use_colnames=True, min_support=0.5)
```

```python
frequent_itemsets = apriori(productdata, min_support=0.1, use_colnames=True )
```

```python
frequent_itemsets['Length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
```

```python
frequent_itemsets
```

```python
target = '{\'bread\'}'
target
```

```python
results_bread = rules[rules['consequents'].astype(str).str.contains(target, na=False)].sort_values(by = 'confidence', ascending = False)
```

```python
results_bread
```

If a customer bought jam, milk and biscuits, or biscuits and sugar, then the customer is more likely to buy bread along wit it.
